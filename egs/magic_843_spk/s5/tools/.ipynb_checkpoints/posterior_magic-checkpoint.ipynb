```python
import sys
import numpy as np
import bisect
import matplotlib.pyplot as plt
from read_files import read_kaldi_post
from read_files import read_kl
from plot_figures import plot_post
```

## Loading the data

```python
root = '/export/b13/xwang/kaldi/kaldi/egs/magic_843_spk/s5'
lat_post_max = root + '/xwang_test/plot_utt_7/lat.1_7.phone.post'
ali_post_max = root + '/xwang_test/plot_utt_7/ali.1_7.phone.post'
lat_ali_kl = root + '/xwang_test/plot_utt_7/kl_1_7.txt'
phone_list = root + '/data/lang/phones.list'
phone_map = root + '/data/lang/map_root_to_phone.map'
lat_post_eps = root + '/lat_post_max_new.eps'
ali_post_eps = root + '/ali_post_max_new.eps'
kl_eps = root + '/post_kl.eps'
epsilon = 10 ** (-10)
```

```python
utt_kl = read_kl.read_kl(lat_ali_kl)

for key in utt_kl.keys():
    kl = np.transpose(utt_kl[key]) 
    lat_post = read_kaldi_post.readKaldiLatPhonePost(lat_post_max, phone_map)
    ali_post = read_kaldi_post.readKaldiLatPhonePost(ali_post_max, phone_map)
    plot_post.plotfigure(lat_post_eps, lat_post[key].T, plot_post.read_phones(phone_list))
    plt.gca().invert_yaxis()
    plt.savefig(lat_post_eps)
    plot_post.plotfigure(ali_post_eps, ali_post[key].T, plot_post.read_phones(phone_list))
    plt.plot(kl, 'm', ms = 6)
    plt.gca().invert_yaxis()
    plt.savefig(ali_post_eps)
    plt.show()
```
